```python
def estimate_alloy_properties_latex_table(alloys, batch_number, liquidus_solidus, folder='Latex\\'):
    """Outputs a latex table with estimated metrics and properties of a given alloy"""

    atomic_percent_latex_shorthand = []
    alloy_densities = []
    alloy_prices = []
    alloy_liquidus = []
    alloy_solidus = []
    alloy_deltaT = []

    for i in range(len(alloys)):
        alloy_liquidus.append(f'{float(liquidus_solidus[2][i]):.0f}')
        alloy_solidus.append(f'{float(liquidus_solidus[4][i]):.0f}')
        alloy_deltaT.append(f'{float(liquidus_solidus[5][i]):.0f}')


    for alloy in alloys:
        atomic_percent_latex_shorthand.append(alloy_calculations.dictionary_to_alloy_latex_shorthand(alloy))

        density = alloy_calculations.estimate_alloy_density(alloy)
        price = alloy_calculations.estimate_alloy_price(alloy)

        alloy_densities.append(f'{density:.2f}')
        alloy_prices.append(f'{price:.2f}')
    
    alloy_index = []
    for i in range(len(alloys)):
        alloy_index.append(f"B{batch_number}-{i+1}")


    estimated_alloy_data = {"\thead{index}": alloy_index,
                            "\thead{Composition \\\ At\%}": atomic_percent_latex_shorthand,
                            "\thead{Price \\\ USD}": alloy_prices,
                            "\thead{Density \\\ g cm$^{-3}$}": alloy_densities,
                            "\thead{T$_{Liquidus}$ \\\ C$^{o}$}": alloy_liquidus,
                            "\thead{T$_{Solidus}$ \\\ C$^{o}$}": alloy_solidus,
                            "\thead{$\Delta$T$_{Liq, Sol}$ \\\ C$^{o}$}": alloy_deltaT
                            }

    estimated_alloy_data = pd.DataFrame(estimated_alloy_data)
    estimated_alloy_data.to_latex(f'{folder}Batch {batch_number}_Estimated Properties.tex', escape=False, index=False, column_format='llrrrrr')

    return estimated_alloy_data
```
